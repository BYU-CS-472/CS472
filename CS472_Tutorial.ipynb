## Loading Data

Each dataset will have a link to a file that can be downloaded. Use curl to download the file and then load the file with whatever you like.

```python
from scipy.io import arff
import pandas as pd

# Download file with curl
!curl https://raw.githubusercontent.com/cs472ta/CS472/master/datasets/voting-dataset.arff --output voting-dataset.arff

voting_data = arff.loadarff('voting-dataset.arff')
voting_df = pd.DataFrame(voting_data[0])

voting_df.head()
```

Data can be moved from a DataFrame to numpy...

```python
import numpy as np

voting_np_data = np.array(voting_df)

voting_np_data
```

...or loaded straight into numpy from scipy.

```python
voting_np_data_from_scipy = np.array(voting_data[0])
voting_np_data_from_scipy
```

Sometimes you will need to convert your data into a different type. Such as going from nominal data to continuous.

```python
from sklearn.preprocessing import LabelEncoder
voting_label_encoded_df = voting_df.apply(LabelEncoder().fit_transform)

voting_label_encoded_df.head()
```

If the data is continuous you can also convert your data with numpy.

```python
!curl https://raw.githubusercontent.com/cs472ta/CS472/master/datasets/linsep2nonorigin.arff --output linsep2nonorigin.arff
    
linsep_data = arff.loadarff('linsep2nonorigin.arff')
linsep_df = pd.DataFrame(linsep_data[0])
linsep_np_data = linsep_df.to_numpy()

print('linsep_np_data dtype:', linsep_np_data.dtype)
print(linsep_np_data)
linsep_np_data = linsep_np_data.astype(float)
print('linsep_np_data dtype:', linsep_np_data.dtype)
print(linsep_np_data)
```

Make sure you monitor your data and you are making the changes you want to make. For example, applying LabelEncoder to a dataset with continuous will also encode all the continuous values

```python
linsep_df.apply(LabelEncoder().fit_transform)
```

## Graphing

We recommend using [matplotlib](https://matplotlib.org/) to create graphs.

```python
import matplotlib.pyplot as plt
import sklearn.datasets as data

X, y = data.load_boston(return_X_y=True)
# Always label your graphs appropriately
plt.title('Boston Housing Prices')
plt.xlabel('Number of rooms')
plt.ylabel('Price in USD')
plt.scatter(X[:, 5], y)
plt.show()
```

```python
# Line graph
x = np.arange(0, 10)
y = x*2
plt.plot(x, y)
plt.show()
```

```python
# Bar graph
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
plt.show()
```

## Tables

You can create tables in [markdown](https://www.markdownguide.org/extended-syntax/#tables) if you want.

| Syntax      | Description |
| ----------- | ----------- |
| Header      | Title       |
| Paragraph   | Text        |



Or you can use [Tabulate](https://pypi.org/project/tabulate/)

```python
# Tabulate also looks pretty good 
from tabulate import tabulate
headers = ["Celestial object", "radius", "mass"]
table = [["Sun",696000,1989100000],["Earth",6371,5973.6],
          ["Moon",1737,73.5],["Mars",3390,641.85]]
print(tabulate(table, headers=headers))
```

You can also display the table in HTML

```python
from IPython.display import HTML, display
display(HTML(tabulate(table, headers=headers, tablefmt='html')))
```
